## 3. ***Data Wrangling***

### Data Wrangling Code

### Step 1 - Data Cleaning

### A. Handling missing values

```python
# Drop rows with missing values in 'name' and 'host_name' columns
df.dropna(subset=['name', 'host_name'], inplace=True)
```

```python
# Fill missing values in 'last_review' and 'reviews_per_month' with appropriate placeholders
df['last_review'].fillna('Not reviewed', inplace=True)
df['reviews_per_month'].fillna(0, inplace=True)
```

```python
# Verify that missing values have been handled
df.isnull().sum()
```

### Removing irrelevant columns

```python
# Drop irrelevant columns
irrelevant_columns = ['id', 'host_id']
airbnb_data_cleaned = df.drop(columns=irrelevant_columns, axis=1)

# Verify the dataset after dropping columns
airbnb_data_cleaned.head()
```

### Step 2 - Standardization

### Standardizing values

```python
# Handle Date-Time Data
# Replace 'Not reviewed' with NaN (missing value)
airbnb_data_cleaned['last_review'] = airbnb_data_cleaned['last_review'].replace('Not reviewed', pd.NaT)

# Convert 'last_review' column to datetime
airbnb_data_cleaned['last_review'] = pd.to_datetime(airbnb_data_cleaned['last_review'])
```

```python
# Print the cleaned dataset
airbnb_data_cleaned.head()
```

### Handling outliers

```python
# Selecting numerical columns
numerical_columns = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']

# Creating boxplots for each numerical column
plt.figure(figsize=(14, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=airbnb_data_cleaned[column])
    plt.title(column)
    plt.tight_layout()

plt.show()
```

```python
# Print the shape of the dataset before removing outliers
print("Shape of dataset before removing outliers:", airbnb_data_cleaned.shape)
```

```python
from scipy import stats

# Calculate z-scores for numerical columns
z_scores = stats.zscore(airbnb_data_cleaned[['price', 'minimum_nights', 'number_of_reviews', 'calculated_host_listings_count', 'availability_365']])

# Define a threshold for identifying outliers (e.g., z-score greater than 3)
threshold = 3

# Filter rows where any z-score exceeds the threshold
outlier_indices = (np.abs(z_scores) > threshold).any(axis=1)

# Remove outliers
airbnb_data_cleaned = airbnb_data_cleaned[~outlier_indices]

# Print the shape of the dataset after removing outliers
print("Shape of dataset after removing outliers:", airbnb_data_cleaned.shape)
```

```python
# Creating boxplots for each numerical column after removing outliers
plt.figure(figsize=(14, 10))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(x=airbnb_data_cleaned[column])
    plt.title(column)
    plt.tight_layout()

plt.show()
```

### Feature Engineering

```python
# Calculate the length of time since the last review (in days)
airbnb_data_cleaned = airbnb_data_cleaned.copy()
airbnb_data_cleaned['days_since_last_review'] = (pd.to_datetime('today') - airbnb_data_cleaned['last_review']).dt.days

# Aggregate information about reviews (e.g., total number of reviews per listing)
airbnb_data_cleaned['total_reviews'] = airbnb_data_cleaned['number_of_reviews'] * airbnb_data_cleaned['reviews_per_month']
```

```python
# Print the head of the data after these steps
airbnb_data_cleaned.head()
```

### What all manipulations have you done and insights you found?

* Handled missing values by dropping rows with missing values in 'name' and 'host_name' columns, and filled missing values in 'last_review' and 'reviews_per_month' columns with appropriate placeholders.
* Standardized date-time data by replacing 'Not reviewed' with NaN and converting the 'last_review' column to datetime format.
* Identified and removed outliers using the z-score method, considering a threshold of 3 standard deviations from the mean for numerical columns such as price, minimum_nights, number_of_reviews, calculated_host_listings_count, and availability_365.
* Conducted feature engineering by calculating the length of time since the last review (in days) and aggregating information about reviews, such as the total number of reviews per listing.

### **What is the average Price of Airbnb Listings over time (by Year)?**

```python
# Extracting the year from the 'last_review' date
airbnb_data_cleaned['review_year'] = airbnb_data_cleaned['last_review'].dt.year

# Grouping the data by review year and calculating the average price
average_price_by_year = airbnb_data_cleaned.groupby('review_year')['price'].mean()

# Plotting the average price over time (by year)
plt.figure(figsize=(10, 6))
plt.plot(average_price_by_year.index, average_price_by_year.values, marker='o', linestyle='-')
plt.title('Average Price of Airbnb Listings Over Time (by Year)')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

I chose a line plot to visualize the average price of Airbnb listings over time (by year) because it effectively shows the trend and changes in price over different years.

Initially, from 2011 to 2014, there was a slight increase in average prices. However, from 2015 onwards, there has been a downward trend in average prices, with some fluctuations in between.

This suggests that there might have been changes in demand, supply, or market dynamics that influenced the pricing trends of Airbnb listings over the years.

Understanding the trend in average prices over time allows Airbnb hosts and stakeholders to adjust their pricing strategies accordingly. They can capitalize on periods of increasing demand by adjusting prices upwards and offer discounts during slower seasons to attract more bookings, thereby maximizing revenue.

If the analysis reveals a consistent upward trend in average prices without corresponding increases in demand, there's a risk of overpricing. Overpricing can lead to decreased bookings, lower occupancy rates, and negative reviews, ultimately impacting the business's reputation and profitability.
